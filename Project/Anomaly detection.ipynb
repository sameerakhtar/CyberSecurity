```python
def eval_EER(genuine_user_scores, other_user_scores):
    labels = [0]*len(genuine_user_scores) + [1]*len(other_user_scores)
    
    
    # f is the false_positive_rates
    # t is the true positive rates
    f, t, thresholds = roc_curve(labels, genuine_user_scores + other_user_scores)
    m_rates = 1 - t
    f_r = f
    diff = m_rates - f_r
    id_1 = np.argmin(diff[diff >= 0])
    id_2 = np.argmax(diff[diff < 0])
    x = [m_rates[id_1], f_r[id_1]]
    y = [m_rates[id_2], f_r[id_2]]
    b = ( x[0] - x[1] ) / ( y[1] - x[1] - y[0] + x[0] )
    eer_ = x[0] + b * ( y[0] - x[0] )
    return eer_


def man_evaluate(subject_id):
        eer_arr = []
        user_scores = []
        impos_scores = []
        mean_vector = []
        
        for id in subject_id:        
            #separating genuine user data and imposter data 
            gen_user_data = data.loc[data.subject == id].iloc[:,3:]
                                        
            impos_data = data.loc[data.subject != id].iloc[:,:]
            
            #--splitting genuine user data into training and testing data
            train_data = gen_user_data[:200]
            
            test_gen = gen_user_data[200:]
            
            #--imposter testing data
            test_impos = impos_data.groupby("subject").head(5).iloc[:, 3:]
                                 
 
            #--manhatten training , calculating mean and  mean absolute deviation 
            mean_vector = train_data.mean().values
            mad_vector  = train_data.mad().values
            
            
            
            #---testing
            
            #--calculatig user scores and imposter scores using test data and manhetten training
            for i in range(test_gen.shape[0]):
                current_scr = 0
                for j in range(len(mean_vector)):
                    current_scr = current_scr + abs(test_gen.iloc[i].values[j] -  mean_vector[j]) / mad_vector[j]                
                user_scores.append(current_scr)

            for i in range(test_impos.shape[0]):
                current_scr = 0
                for j in range(len(mean_vector)):
                                current_scr = current_scr +  abs(test_impos.iloc[i].values[j] -  mean_vector[j]) / mad_vector[j]
                impos_scores.append(current_scr)
            
            #--calculating EER from user and imposter scores and appending it into eer_array
            eer_arr.append(eval_EER(user_scores, impos_scores))
            
        #return mean of array and standard deviation    
        return np.mean(eer_arr) , np.std(eer_arr)
   
```

```python
uniq_subject_id = data["subject"].unique()

print('Manhatten Average EER and standard devistion are :',man_evaluate(uniq_subject_id))
```
